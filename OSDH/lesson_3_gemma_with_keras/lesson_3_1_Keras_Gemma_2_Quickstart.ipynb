## Quickstart

```
import keras_nlp
import keras

# Run at half precision.
keras.config.set_floatx("bfloat16")

# using 9B base model
gemma_lm = keras_nlp.models.GemmaCausalLM.from_preset("gemma2_9b_en")
gemma_lm.summary()
```

```
result = gemma_lm.generate("It was a dark and stormy night.", max_length=256)
print(result)
```
